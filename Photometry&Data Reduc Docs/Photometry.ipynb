```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
from photutils import DAOStarFinder
from photutils import aperture_photometry, CircularAperture, CircularAnnulus
```

```python
## FINDING SOURCES IN AN IMAGE ##
# For this portion I used example code from photutils.readthedocs to help.

# Getting the data for the image in array format
image_data = fits.getdata("quadRU.fits").astype(np.int32)

# Retrieving and displaying the data array shape for reference
print("Image shape:")
print(image_data.shape)
print()

# Finding source objects within the image and putting them into a table
daofind = DAOStarFinder(fwhm = 4, threshold = 100)
sources = daofind(image_data)

# Viewing the table of all found sources
for col in sources.colnames:
    sources[col].info.format = '%.8g'
sources
```

```python
## REMOVING SOURCES TOO CLOSE TO ANY EDGE OF THE IMAGE (< 10 PIXELS) ##

# Retrieving columns from the sources table
ids = sources['id']
xpos = sources['xcentroid']
ypos = sources['ycentroid']
mag = sources['mag']

# If a source object in the sources table is positioned within 10 pixels of any edge of the image, append its
# row number to a list and print its row number and ID
rows_to_remove = []

print("Rows to remove (and corresponding IDs):")
for row in range(0, xpos.size):
    if ((xpos[row] < 10) or (ypos[row] < 10) or (xpos[row] > 1017) or (ypos[row] > 1013)):
        print("row = " + str(row) + ", ID = " + str(ids[row]))
        rows_to_remove.append(row) 
print()

# Make a copy of the table
sources_final = sources.copy()

# Removing the 'bad' rows from the sources table    
sources_final.remove_rows(rows_to_remove)

# Viewing the updated table - after removal of the 'bad' rows
print("Source Information After Removing Objects Within 10 Pixels of Each Edge")
for col in sources_final.colnames:
    sources_final[col].info.format = '%.8g'
sources_final
```

```python
## PLOTTING STARS FOUND ON TOP OF THE IMAGE ##

plt.imshow(image_data, vmin = np.percentile(image_data, 5), vmax = np.percentile(image_data, 90),
           origin = 'lower', cmap = 'binary_r', interpolation = 'nearest');
plt.plot(sources_final['xcentroid'], sources_final['ycentroid'], 'y+');
plt.title('Positions of Stars Detected by DAOStarFinder');
```

```python
## CREATING CIRCULAR APERTURES ##

positions = (sources_final['xcentroid'], sources_final['ycentroid'])
circular_apertures = CircularAperture(positions, r = 10)
phot_table_circ = aperture_photometry(image_data, circular_apertures, method = 'exact')

# Viewing the photometry table
for col in phot_table_circ.colnames:
    phot_table_circ[col].info.format = '%.8g'
phot_table_circ
```

```python
## PLOTTING THE CIRCULAR APERTURES ON TOP OF THE IMAGE ##

plt.imshow(image_data, vmin = np.percentile(image_data, 5), vmax = np.percentile(image_data, 90),
           origin = 'lower', cmap = 'binary_r', interpolation = 'nearest')
circular_apertures.plot(color = 'red', lw = 1.5, alpha = 0.5)
```

```python
## CREATING CIRCULAR ANNULI ##

annuli = CircularAnnulus(positions, r_in = 15, r_out = 17)
phot_table_ann = aperture_photometry(image_data, annuli, method = 'exact')

# Viewing the photometry table for annuli
for col in phot_table_ann.colnames:
    phot_table_ann[col].info.format = '%.8g'
phot_table_ann
```

```python
## PLOTTING THE ANNULI ON TOP OF THE IMAGE ##

plt.imshow(image_data, vmin = np.percentile(image_data, 5), vmax = np.percentile(image_data, 90),
           origin = 'lower', cmap = 'binary_r', interpolation = 'nearest')
annuli.plot(color = 'cyan', lw = 1.0, alpha = 0.5)
```
