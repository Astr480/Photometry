```python
bkg_estimator = SExtractorBackground()
sigma_clip = SigmaClip(sigma=2.5)


bkg = Background2D(quad_data, (50, 50), filter_size=(2, 2),sigma_clip=sigma_clip,
                   bkg_estimator=bkg_estimator)


print(bkg.background_median)
print(bkg.background_rms_median)
```

```python
#plot of the background, calculated above

fig = plt.figure()
fig.set_size_inches(14,6)
fig.tight_layout()

plt.imshow(bkg.background, origin='lower', cmap='Greys_r')
```

```python
quad_clean = quad_data - bkg.background
```

```python
mean, median, std = sigma_clipped_stats(quad_data, sigma=3.0)
print((mean, median, std))
```

```python
daofind = DAOStarFinder(fwhm=4.0, threshold=5.0*std)
sources = daofind(quad_data)

for col in sources.colnames:
    sources[col].info.format = '%.8g'

print('Number of stars found:', len(sources))
print('')
print(sources)
```

```python
positions = (sources['xcentroid'], sources['ycentroid'])    
apertures = pu.CircularAperture(positions, r=4.)
```

```python
quad_table = aperture_photometry(quad_data, apertures)
quad_table['aperture_sum'].info.format = '%.8g'  # for consistent table output
print(quad_table)
```

```python
annulus_apertures = CircularAnnulus(positions, r_in=6., r_out=8.)

apers = [apertures, annulus_apertures]
quad_table2 = aperture_photometry(quad_data, apers)
for col in quad_table2.colnames:
    quad_table2[col].info.format = '%.8g'  # for consistent table output
print(quad_table2)
```

```python
fig = plt.figure()
fig.set_size_inches(14,6)
fig.tight_layout()

plt.subplot(1, 2, 1)
plt.imshow(quad_data, cmap='Greys', origin='lower', figure=annulus_apertures.plot(color='blue', lw=1.5, alpha=0.5))
plt.title('Annulus')

plt.subplot(1 ,2, 2)
plt.imshow(quad_data, cmap='Greys', origin='lower', figure=apertures.plot(color='red', lw=1.5, alpha=0.5))
plt.title('Aperture')
```

```python
bkg_mean = quad_table2['aperture_sum_1'] / annulus_apertures.area()
bkg_sum = bkg_mean * apertures.area()
final_sum = quad_table2['aperture_sum_0'] - bkg_sum
```

```python
annulus_values = final_sum
aperture_values = quad_table['aperture_sum']

x = annulus_values
y = aperture_values

plt.scatter(x,y)
plt.xlim(left=0,right=400000)
plt.ylim(bottom=0,top=500000)
plt.title('Annuli vs. Apertures')
plt.xlabel('Annulus data')
plt.ylabel('Aperature data')
```
